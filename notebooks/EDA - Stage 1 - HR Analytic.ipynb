# **B. Mengolah Dataset in_time dan out_time**

## Mendefinisakan column awal yang Kosong dengan `EmployeeID`

```python
in_time = in_time.rename(columns={'Unnamed: 0': 'EmployeeID'})
out_time = out_time.rename(columns={'Unnamed: 0': 'EmployeeID'})
```

## Melakukan Copy Data dan Membuat data Jam Kerja (work_hours)

```python
in_dt = in_time.copy()
out_dt = out_time.copy()

for col in in_dt.columns[1:]:
    in_dt[col] = pd.to_datetime(in_dt[col])
for col in out_dt.columns[1:]:
    out_dt[col] = pd.to_datetime(out_dt[col])

work_hours = (out_dt.iloc[:, 1:] - in_dt.iloc[:, 1:]).apply(
    lambda x: x.dt.total_seconds() / 3600
)
```

## Melakukan Penambahan Fitur Baru Rata-rata Jam Kerja `AvgWorkHours`

```python
avg_work_hours = work_hours.mean(axis=1)

work_hours = pd.concat(
    [in_dt[['EmployeeID']], avg_work_hours.rename('AvgWorkHours')],
    axis=1
).copy()
```

## Imputasi Missing Value

```python
# Imputasi Missing Value
num_cols = [
    'Age','DistanceFromHome','Education','MonthlyIncome','NumCompaniesWorked',
    'PercentSalaryHike','StandardHours','StockOptionLevel','TotalWorkingYears',
    'TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager',
    'EnvironmentSatisfaction','JobSatisfaction','WorkLifeBalance','JobInvolvement',
    'PerformanceRating','AvgWorkHours'
]
cat_cols = [c for c in df.columns if c not in ['Attrition','AttritionFlag'] and not pd.api.types.is_numeric_dtype(df[c])]

for c in num_cols:
    df[c] = df[c].fillna(df[c].median())
for c in cat_cols:
    mode = df[c].mode(dropna=True)
    df[c] = df[c].fillna(mode.iloc[0] if len(mode)>0 else 'Unknown')

df.info()
```

Alasannya winsorisasi tidak membuang observasi, hanya menyesuaikan nilai ekstrem ke batas yang lebih realistis

## Handling Outlier

```python
# --- Winsorisasi Outliers (IQR) untuk variable kontinyu utama ---
cont_cols = ['MonthlyIncome','DistanceFromHome','TotalWorkingYears',
             'YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager','AvgWorkHours']
for c in cont_cols:
    if c in df.columns:
        q1, q3 = df[c].quantile(0.25), df[c].quantile(0.75)
        iqr = q3 - q1
        lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
        df[c] = df[c].clip(lower=lower, upper=upper)
```

```python
plt.figure(figsize=(15, 20))
num_plots = len(cont_cols)
cols = 3
rows = (num_plots + cols - 1) // cols

for i in range(0, num_plots):
    plt.subplot(rows, cols, i + 1)
    sns.boxplot(y=df[cont_cols[i]], orient='v')
    plt.title(f'Boxplot of {cont_cols[i]}')

plt.tight_layout()
plt.show()
```

# F. Feature engineering (one-hot encoding, feature scaling, feature extraction)

### Feature Extraction

```python
# Fitur Tambahan berbasis jam kerja ---
df['OvertimeRatio'] = df['AvgWorkHours'] / df['StandardHours'].replace(0, np.nan)
df['HourGapFromStandard'] = df['AvgWorkHours'] - df['StandardHours']
df['OvertimeFlag'] = (df['AvgWorkHours'] > df['StandardHours']).astype(int)
df['WorkloadIndex'] = df['AvgWorkHours'] / (1 + df['JobLevel'].replace(0, np.nan))
df['WLB_Mismatch'] = (df['AvgWorkHours'] - df['StandardHours']) * (5 - df['WorkLifeBalance'])
```

```python
display(df[['EmployeeID', 'AttritionFlag','OvertimeRatio','HourGapFromStandard', 'OvertimeFlag', 'WorkloadIndex', 'WLB_Mismatch']])
```

```python
# Fitur tambahan untuk melihat Aspek Kompensasi & Karier ---
df['IncomePerHour'] = df['MonthlyIncome'] / (4.33 * df['AvgWorkHours'].replace(0, np.nan))
df['TimeSinceLastPromoRatio'] = (df['YearsSinceLastPromotion'] / df['YearsAtCompany'].replace(0, np.nan)).fillna(0)
df['ManagerStabilityIndex'] = (df['YearsWithCurrManager'] / df['YearsAtCompany'].replace(0, np.nan)).fillna(0)

# Handle division by zero for CareerMobility more robustly
# Replace 0 in TotalWorkingYears with NaN to avoid division by zero which would produce inf or NaN
temp_total_working_years = df['TotalWorkingYears'].replace(0, np.nan)
df['CareerMobility'] = df['NumCompaniesWorked'] / temp_total_working_years

# Explicitly replace any resulting inf or -inf values with NaN, then fill all NaNs with 0
df['CareerMobility'].replace([np.inf, -np.inf], np.nan, inplace=True)
df['CareerMobility'].fillna(0, inplace=True)
```

```python
display(df[['EmployeeID', 'IncomePerHour', 'TimeSinceLastPromoRatio', 'ManagerStabilityIndex', 'CareerMobility']])
```

```python
df[['EmployeeID', 'IncomePerHour', 'TimeSinceLastPromoRatio', 'ManagerStabilityIndex', 'CareerMobility']].info()
```

```python
# Fitur Tambahan Terkait Survey Kepuasan * jam kerja & commute
sat_cols = [c for c in ['JobSatisfaction','EnvironmentSatisfaction','RelationshipSatisfaction'] if c in df.columns]
if sat_cols:
    df['Satisfaction_X_Hours'] = df[sat_cols].sum(axis=1) * df['AvgWorkHours']
df['CommuteStressIndex'] = np.log1p(df['DistanceFromHome']) * df['AvgWorkHours']
```

```python
# --- Tenure bucket ---
bins = [0,1,3,5,10,np.inf]; labels = ['0-1','1-3','3-5','5-10','>10']
df['TenureBucket'] = pd.cut(df['YearsAtCompany'], bins=bins, labels=labels, right=False)
```

```python
display(df[['EmployeeID','Satisfaction_X_Hours', 'CommuteStressIndex', 'TenureBucket']])
```

```python
# Tambahan Fitur Risk Score dan Risk Cohort
dept_median_income_per_hour = df.groupby('Department')['IncomePerHour'].transform('median')
dept_q75_commute = df.groupby('Department')['CommuteStressIndex'].transform(lambda s: s.quantile(0.75))

risk = (
    (df['OvertimeRatio'] > 1.0).astype(int) * 1.0 +
    (df['WorkLifeBalance'] <= 2).astype(int) * 1.0 +
    (df['TimeSinceLastPromoRatio'] > 0.5).astype(int) * 0.5 +
    (df['IncomePerHour'] < dept_median_income_per_hour).astype(int) * 0.5 +
    (df['CommuteStressIndex'] > dept_q75_commute).astype(int) * 0.5
)
df['RiskScore'] = risk

def bucketize(score):
    if pd.isna(score): return 'Unknown'
    if score >= 2.5:   return 'High'
    elif score >= 1.5: return 'Medium'
    else:              return 'Low'
df['RiskCohort'] = df['RiskScore'].apply(bucketize)
```

# **G. Data Standarization**

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, RobustScaler

# Load data dari jalur yang benar dan dengan nama file yang benar
df_new_feature = pd.read_csv('../data/interim/hr_analytics_new_feature.csv')

# --- Global Cleaning: Tangani semua nilai tak terhingga dan NaN ---
# Dapatkan semua kolom numerik yang akan diteruskan ke preprosesor
numerical_cols_for_preprocess = df_new_feature.select_dtypes(include=np.number).columns.tolist()
numerical_cols_for_preprocess = [c for c in numerical_cols_for_preprocess if c not in ['EmployeeID', 'AttritionFlag']] # Exclude ID and target for imputation purposes

for col in numerical_cols_for_preprocess:
    # Ganti inf dan -inf dengan NaN
    df_new_feature[col] = df_new_feature[col].replace([np.inf, -np.inf], np.nan)
    # Isi nilai NaN yang tersisa dengan median kolom tersebut
    # Gunakan .median() hanya jika ada nilai selain NaN; jika tidak, isi dengan 0
    if df_new_feature[col].isnull().any():
        median_val = df_new_feature[col].median()
        df_new_feature[col].fillna(median_val if not pd.isna(median_val) else 0, inplace=True)
```

```python
# Pastikan ada target numerik
if 'AttritionFlag' not in df_new_feature.columns and 'Attrition' in df_new_feature.columns:
    df_new_feature['AttritionFlag'] = df_new_feature['Attrition'].map({'Yes':1,'No':0}).astype('Int64')
```

```python
# Deteksi kolom kategorikal (object), kecuali Attrition
cat_cols = [c for c in df_new_feature.columns if df_new_feature[c].dtype == 'object']
cat_cols = [c for c in cat_cols if c not in ['Attrition']]
```

```python
# Fitur numerik (skew)
num_skew = [
    'MonthlyIncome','DistanceFromHome','AvgWorkHours','IncomePerHour',
    'CommuteStressIndex','OvertimeRatio','HourGapFromStandard',
    'WorkloadIndex','WLB_Mismatch','TimeSinceLastPromoRatio'
]
num_skew = [c for c in num_skew if c in df_new_feature.columns]
```

```python
# Log1p untuk MonthlyIncome sebelum scaling
df_new_feature['log1p_MonthlyIncome'] = np.log1p(pd.to_numeric(df_new_feature['MonthlyIncome'], errors='coerce'))
num_skew = ['log1p_MonthlyIncome' if c=='MonthlyIncome' else c for c in num_skew]
```

```python
# Pilih scaler: Robust untuk fitur dengan outlier kuat

# Semua kolom numerik akan diskalakan (kecuali kolom ID, target, dan original/redundant income columns)
numerical_cols_all = df_new_feature.select_dtypes(include=np.number).columns.tolist()
numerical_cols_all = [c for c in numerical_cols_all if c not in ['EmployeeID', 'AttritionFlag', 'MonthlyIncome', 'log_income', 'StandardHours']]

# Pisahkan Fitur yang skewed (already in standard_cols/robust_cols) dan non-skewed
num_skewed_handled = standard_cols + robust_cols
num_non_skewed_cols = [c for c in numerical_cols_all if c not in num_skewed_handled]

preprocess = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols),
        ('std_skew',    StandardScaler(), standard_cols), # Scaler for skewed but not robust
        ('rob_skew',    RobustScaler(),   robust_cols),   # RobustScaler for highly skewed
        ('std_non_skew', StandardScaler(), num_non_skewed_cols) # Scaler for other numerical features
    ],
    remainder='drop'
)

X = preprocess.fit_transform(df_new_feature)
```

```python
# Nama kolom keluaran onehot
onehot = preprocess.named_transformers_['onehot']
onehot_feature_names = list(onehot.get_feature_names_out(cat_cols)) if cat_cols else []

# Gabungkan semua nama kolom yang telah diproses
out_cols = onehot_feature_names + standard_cols + robust_cols + num_non_skewed_cols

X_df = pd.DataFrame(X, columns=out_cols)
```

```python
# Tambahkan target dan identifier
for col in ['AttritionFlag','EmployeeID']:
    if col in df_new_feature.columns:
        X_df[col] = df_new_feature[col].values
```
